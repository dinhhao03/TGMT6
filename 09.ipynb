```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
my_path = 'C:/Users/nguye/images/01/img4.jpg'
```

```python
# Đường dẫn ảnh
img = Image.open(my_path)
plt.imshow(img)
```

```python
# Chuyển sang ảnh xám
img_gray = img.convert('L')
plt.imshow(img_gray, cmap='gray')
```

```python
img_array = np.array(img_gray)
print(img_array)
```

```python
plt.figure()
plt.hist(img_array.flatten(), bins=128)
plt.title('Biểu đồ Histogram của hình ảnh')
plt.xlabel('Giá trị của pixel')
plt.ylabel('Số lượng')
plt.show()
```

```python
plt.figure()
plt.gray()
plt.contour(img_gray, origin="image")
plt.show()
```
